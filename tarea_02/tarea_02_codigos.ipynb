**Problema 1. Clasificador de temperatura**

Crea un programa que imprima los números desde -30 hasta 50.

Para cada número, además de imprimirlo, el programa debe mostrar un texto según estas reglas:

si el número es menor que 0, mostrar "Muy frío"
si el número está entre 0 y 24, mostrar "Templado"
si el número es mayor o igual a 25, mostrar "Caluroso"

```python
# lista desde -30 al 50
lista_numeros = range(-30,50)
print(lista_numeros)

for x in range(-30,51):
  if(x < 0):
    print(x, "Muy frío")
  elif(x >= 0 and x <= 24):
    print(x, "Templado")
  elif(x >= 25):
    print(x, "Caluroso")
```

Código que se trabajó en clases. Se enseñó  el uso del print, for, range, if, elif y else.

Se incluye el rango que va desde el -30 al 50, y se imprime con print para que aparezca en pantalla. Se debe escribir hasta el 51.

Aquí es donde aparecen las condiciones por donde tienen que recorrer los números: if (si...) x (el número) es menor a 0, entonces se imprime "muy frio"; si es mayor o igual a 0 y menor o igual a 24, se imprime "templado"; si no cumplió con las anteriores y es igual o mayor a 25, entonces se imprime "caluroso".

**Problema 2. Números especiales del robot**

Crea un programa que imprima los números del 1 al 100 que sean divisibles por 5.

Además, el programa debe cumplir estas condiciones:

si el número es divisible por 5, imprimir el número junto con el texto "Bip" si el número también es divisible por 10, imprimir el número junto con el texto "Bip Bip" Ejemplo de salida

```python
# numero desde 1 al 100 divisibles por 5
lista_numeros = range(1,101)
print(lista_numeros)

for x in lista_numeros:
  if(x % 5 == 0):
    if(x % 10 == 0):
      print(x, "Bip Bip")
    else:
     print(x, "Bip")
```

En range, se incluye la lista de números del 1 al 100, donde se escribe 101 porque el programa no lo cuenta, y se imprime con print para que aparezca en pantalla.

Se usa esta fórmula (x % 5 == 0) para decir si x es divisible por 5, mismo caso que con el 10.

Si es divisible por 5, entonces pasa a la siguiente condición. Si, además, se puede dividir por 10, se imprime con el texto "bip bip". Si solo es dividible por 5, entra en la instrucción de else y se imprime con el texto "bip".

**Problema 3. Detector de puntaje gamer**

Crea un programa que solicite al usuario dos números, los sume y luego muestre un mensaje según el resultado.

Reglas

si la suma es menor que 50, mostrar "Nivel principiante".
si la suma es mayor o igual a 50 y menor que 80, mostrar "Nivel intermedio".
si la suma es mayor o igual a 80, mostrar "Nivel pro".

```python
# Detector de puntaje gamer

numero_1 = int(input("Ingrese el primer numero: "))
numero_2 = int(input("Ingrese el segundo numero: "))

suma = numero_1 + numero_2
print("La suma es: ", suma)

if(suma < 50):
  print("Nivel principiante")
elif(suma >= 50 and suma < 80):
  print("Nivel intermedio")
elif(suma >= 80):
  print("Nivel pro")
```

Como pide dos números, se debe anotar numero_1 y numero_2, porque no sabemos cuál es ese número. Solicita que el usuario ingrese esos números, por lo que la instrucción de int hace referencia a un número entero, y el input permite que el usuario lo ingrese.

En el número 6 y 7, se pide que el numero_1 y numero_2 se sumen, y con el print permite que se muestre en pantalla.

A partir del número 9, se empieza a aplicar las condiciones y reglas.
Si la suma es menor a 50, se imprime el texto "nivel principiante", ignorando las siguientes condiciones y finalizando su lectura, por así decirlo. Esto pasa igual con las siguientes reglas.

**Problema 4. Calculador de perfil musical**

Crea un programa que evalúe el perfil del usuario según su edad y sus gustos.

El usuario debe recibir los siguientes parámetros:

edad
le_gusta_musica
le_gusta_bailar
Y debe retornar uno de los siguientes mensajes:

"Eres una estrella de la fiesta" si la persona tiene 18 o más años, le gusta la música y le gusta bailar.
"Te gusta la música, pero no tanto bailar" si le gusta la música pero no le gusta bailar.
"Tienes tu propio estilo" en cualquier otro caso.

```python
# Calculador de perfil musical

edad = int(input("Ingrese su edad: "))
le_gusta_musica = input("Ingrese si le gusta la musica (si/no): ")
le_gusta_bailar = input("Ingrese si le gusta bailar (si/no): ")

if(edad >= 18 and le_gusta_musica == "si" and le_gusta_bailar == "si"):
  print("Eres una estrella de la fiesta")
elif(le_gusta_musica == "si" and le_gusta_bailar == "no"):
  print("Te gusta la música, pero no tanto bailar")
else:
  print("Tienes tu propio estilo")
```

Se hacen tres casillas con int(input( más los parámetros edad le_gusta_musica, le_gusta_bailar, para que el usuario escriba su edad (número entero) y en las últimas dos "si" o "no".

A partir del 7 se dan instrucciones con condiciones: if -si tal cosa- la edad es mayor o igual a 18, y escribió "si" en le_gusta_musica y "si" en le_gusta_bailar, entonces se imprime el texto "eres una estrella de la fiesta". Si no cumple con una, dos o cualquiera de esas tres condiciones, entonces va a la siguiente condición. Esto pasa igual hasta llegar a la línea 11 con else, porque si el perfil no encajó en cualquiera de las anteriores opciones, se queda en este último y se imprime "tienes tu propio estilo".
